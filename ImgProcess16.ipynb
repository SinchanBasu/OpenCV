```python
# Python | Visualizing image in different color spaces
# Python program to read image as RGB

import cv2
import matplotlib.pyplot as plt  
img = cv2.imread('input7.png')
plt.imshow(img)
```

```python
# Python program to read image as GrayScale
import cv2
img = cv2.imread('input7.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Python program to read image as YCrCb Color Space
import cv2
img = cv2.imread('input7.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)
plt.imshow(img)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Pyrhon Program to read image as HSV Color Space
import cv2
img = cv2.imread('input7.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
plt.imshow(img)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Python program to read image as LAB color space
import cv2
img = cv2.imread('input7.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2LAB)
plt.imshow(img)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Python program to read image as Edge Map
import cv2
img = cv2.imread('input7.png')
laplacian = cv2.Laplacian(img,cv2.CV_64F)
cv2.imshow('EdgeMap',laplacian)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
