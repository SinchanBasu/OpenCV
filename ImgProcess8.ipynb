```python
# OpenCV Python Program to analyze an image using Histogram
import cv2
import matplotlib.pyplot as plt
img = plt.imread('flower.jpg',0)
histr = cv2.calcHist([img],[0],None,[256],[0,256])
plt.plot(histr)
plt.show()
```
